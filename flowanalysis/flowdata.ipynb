```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#import csv as df
df = pd.read_csv('floreadastats (2).csv')
df = df.dropna()
```

```python
df['pseudoPSI'] = df['Mean PE-A'] / (df['Mean PE-A'] + df['Mean FITC-A'])
df
```

```python
g1df = df[(df['Guide'] == 1) | (df['Guide'] == 0)].copy()
g1df = g1df.dropna()
g1df
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

labels = g1df['Design']
averages = []
errors = []
data_points = []
labelList = []

for label in labels:
    if label not in labelList:
        label_data = g1df[g1df['Design'] == label]['Mean PE-A']
        averages.append(label_data.mean())
        errors.append(label_data.std())  # Standard deviation as error
        data_points.append(label_data.values)
        labelList.append(label)
# Create the plot
plt.figure(figsize=(10, 6))
x_pos = np.arange(len(labelList))

# Create bars with error bars
bars = plt.bar(x_pos, averages, yerr=errors, capsize=5, 
               alpha=0.7, label='Average  STD')

# Add individual data points with jitter to avoid overlap
for i, points in enumerate(data_points):
    # Add small random jitter to x-position
    jitter = np.random.normal(0, 0.05, size=len(points))
    plt.scatter(i + jitter, points, color='black', alpha=0.6, 
                s=30, zorder=3, label='Individual points' if i == 0 else "")

plt.xlabel('Designs')
plt.ylabel('Mean PE-A')
plt.title('Guide 1')
plt.xticks(x_pos, labelList)
plt.legend()
plt.tick_params(axis='x', labelrotation=45)

plt.tight_layout()
plt.show()
```
